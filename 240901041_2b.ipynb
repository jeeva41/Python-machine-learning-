```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
grades = np.random.randint(40, 100, size=(30,3))
print("Student Grades (Math, Science, English):\n")
print(grades)
subjects = ["Math", "Science", "English"]
avg = np.mean(grades, axis=0)
highest = np.max(grades, axis=0)
lowest = np.min(grades, axis=0)
print("\nSubject-wise Statistics")
for i in range(3):
    print(f"{subjects[i]} -> Average: {avg[i]:.2f}, Highest: {highest[i]}, Lowest: {lowest[i]}")
total_scores = np.sum(grades, axis=1)
avg_scores = np.mean(grades, axis=1)
top_student = np.argmax(total_scores)
print("\nTop Performing Student")
print("Student Index:", top_student)
print("Total Marks:", total_scores[top_student])
print("Average Marks:", avg_scores[top_student])
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.hist(grades[:,0], bins=10, color='blue')
plt.title("Math Grades")
plt.subplot(1,3,2)
plt.hist(grades[:,1], bins=10, color='blue')
plt.title("Science Grades")
plt.subplot(1,3,3)
plt.hist(grades[:,2], bins=10, color='blue')
plt.title("English Grades")
plt.tight_layout()
plt.show()
```
