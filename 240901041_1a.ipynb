```python
import numpy as np
import pandas as pd
roll_no = np.arange(101, 121)
scores = np.array([85,74,25,46,75,89,76,63,46,52,43,74,42,81,73,68,57,39,79,81])
df = pd.DataFrame({
    "Roll No": roll_no,
    "Score": scores
})
average = np.mean(scores)
highest = np.max(scores)
lowest = np.min(scores)
print("Average Score:", average)
print("Highest Score:", highest)
print("Lowest Score:", lowest)
df["Result"] = df["Score"].apply(lambda x: "Pass" if x >= 50 else "Fail")
def grade(score):
    if score >= 85:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 50:
        return "D"
    else:
        return "F"
df["Grade"] = df["Score"].apply(grade)
print("\nStudent Result Table")
print(df)
```
