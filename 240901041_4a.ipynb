```python
import pandas as pd
import numpy as np

# Sample EEG dataset (Participant, Electrode readings)
data = {
    "Participant": ["P1","P1","P1","P2","P2","P2","P3","P3","P3"],
    "Electrode": ["E1","E2","E3","E1","E2","E3","E1","E2","E3"],
    "Brainwave": [18.5, 11.3, 11.2, 14.1, 13.8, 14.6, 12.7, 17.9, 16.2]
}
df = pd.DataFrame(data)

print("EEG Dataset:")
print(df)
print("\nMean Brainwave Activity:")
print(df.groupby(["Participant","Electrode"])["Brainwave"].mean())
print("\nVariance of Brainwave Activity:")
print(df.groupby(["Participant","Electrode"])["Brainwave"].var())
print("\nPeak Brainwave Values:")
print(df.groupby(["Participant","Electrode"])["Brainwave"].max())
trend = df.groupby("Participant")["Brainwave"].mean()

print("\nAverage Brainwave Activity per Participant:")
print(trend)
```
