```python
import numpy as np
import pandas as pd
steps = np.array([6000,5233,8000,4000,7500,
                  9600,4100,8200,6200,7200,
                  3000,4100,8500,9200,10000,
                  6900,7100,7700,8300,5400,
                  8800,9400,6800,7200,8450,
                  8410,9400,7300,6200,5900]
)

days = np.arange(1, 31)
df = pd.DataFrame({
    "Day": days,
    "Step_Count": steps
})
total_steps = np.sum(steps)
average_steps = np.mean(steps)
max_steps = np.max(steps)
min_steps = np.min(steps)

print("Total Steps:", total_steps)
print("Average Steps per Day:", average_steps)
print("Maximum Steps:", max_steps)
print("Minimum Steps:", min_steps)

high_activity = df[df["Step_Count"] > 8000]

print("\nHigh Activity Days:")
print(high_activity)

df["Trend"] = df["Step_Count"].diff()

print("\nStep Count Trend:")
print(df)
```
