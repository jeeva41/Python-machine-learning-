```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
'Price': [6800, 1300, 600, 350, 150],
'Quantity': [10, 15, 20, 25, 50]
}
df = pd.DataFrame(data)

print("Product Inventory Data:\n")
print(df)
df['Total_Value'] = df['Price'] * df['Quantity']
print("\nInventory with Total Value:")
print(df)
print("\nAverage Product Price:", df['Price'].mean())

print("\nMost Expensive Product:")
print(df.loc[df['Price'].idxmax()])
print("\nProduct with Highest Quantity:")
print(df.loc[df['Quantity'].idxmax()])
sorted_df = df.sort_values(by='Price')
print("\nProducts Sorted by Price:")
print(sorted_df)
df.plot(x='Product', y='Quantity', kind='bar', title="Product Quantity Distribution")
plt.ylabel("Quantity")
plt.show()
```
