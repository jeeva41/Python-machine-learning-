```python
import numpy as np
import pandas as pd

# Component data (example values)
data = {
    "Component": ["R1", "L1", "C1"],
    "Impedance": [10+0j, 0+5j, 0-3j]   # Complex impedance values
}

df = pd.DataFrame(data)
print("Circuit Components:")
print(df)
Z = np.array(df["Impedance"])
Z_series = np.sum(Z)

print("\nTotal Impedance (Series Connection):")
print(Z_series)
Z_parallel = 1 / np.sum(1 / Z)

print("\nTotal Impedance (Parallel Connection):")
print(Z_parallel)
```
