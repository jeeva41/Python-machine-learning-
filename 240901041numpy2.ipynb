```python
import numpy as np
arr=np.array([[-1,2,0,4],[4,-0.5,6,0],[2,6,0.7,8],[3,-7,4,20]])
print("Orignal array:\n",arr)
```

```python
print('\n Every other rows:\n',arr[0:3:2])
```

```python
arr=np.array([1,2,3,4,5,6,7,])
print("\n original array",arr)
print("\n Return every other elements in the array:arr[::2]:",arr[::2])
```

```python
#slicing array
arr=np.array([[-1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2.0]])
temp=arr[:2,:3]
print("\nArray with first 2 rows and 3 colums ;\n",temp)
```

```python
t=arr[[0,1,2,3],[3,2,1,0]]
print("\nElements at indices(0,3),(1,2),(2,1),(3,0)\n",t)
```

```python
#boolean array indexing
cond=arr>2
#cond is a boolean array
t=arr[cond]
print("\n Elements greater than 2:\n",t)
```

```python
#joining two array
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.concatenate((arr1,arr2))
print(arr)
```

```python
#horizontal join
arr3=np.hstack((arr1,arr2))
print(arr3)
```

```python
#vertical join
arr4=np.vstack((arr1,arr2))
print(arr4)
```

```python
#depth join
arr5=np.dstack((arr1,arr2))
print(arr5)
```

```python
#spillting of array
arr6=np.array([1,2,3,4,5,6])
arr6=np.array_split(arr,3)
print(arr)
print(arr6)
print(arr6[0])
print(arr6[1])
print(arr6[2])
```
