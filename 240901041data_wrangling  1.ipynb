```python
dfl=pd.DataFrame([[11,12],[3,4]])
df2=pd.DataFrame([[5,6],[7,8]])
def myfunc(a, b):
    if(a.sum()>b.sum()):
        return a
    else:
        return b
print(dfl)
print(df2)
print(dfl.combine(df2,myfunc))
```

```python
import pandas as pd
d1 = {
    "name":["sai","sub","poo"],
    "age":[20,21,22]
}
d2 = {
    "name":["sai","sub","poo"],
    "qualified": [True, True, False]
}
dfl=pd.DataFrame(d1)
print("DF1:\n",dfl)
df2=pd.DataFrame(d2)
print("DF2:\n",df2)
newdf1=dfl.merge(df2, on="name", how='left')
```

```python
newdf2=df2.merge(dfl, on="name", how='right')
print("Left Merge:\n",newdf1)
print("Right Merge:\n",newdf2)
```

```python
# Create DataFrame
df=pd.DataFrame({'team':['A','B','C','D'],
'points':[90,56,78,88],
'assists':[12,34,23,44],
'rebounds':[56,43,55,12]})
# View DataFrame
print(df)
#reshape DataFrame from wide format to long format
df=pd.melt(df,id_vars='team', value_vars=['points','assists','rebounds'])
#View updated DataFrame
print("\nReshaped DataFrame:\n",df)
```
